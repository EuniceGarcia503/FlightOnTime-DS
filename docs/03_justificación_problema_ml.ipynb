#**Justificación del problema de Machine Learning.**

##**1. El problema de negocio requiere Machine Learning.**

##**1.1. Naturaleza multifactorial del retraso operativo.**

**Multifactorial** significa que no hay una sola causa que explique el vuelo.

Un vuelo puede retrasarse por una combinación de condiciones entre las más comunes:

*   Aeropuerto de origen.
*   Aerolínea.
*   Hora del día.
*   Día de la semana.
*   Mes del año.
*   Estacionalidad (otoño,invierno,verano,primavera).
*   Clima.

Por eso no basta con una regla simple. Se necesita un modelo que pueda aprender **cómo se combinan esos factores.**

##**1.2. Problema no determinista.**

**No determinista** significa que **con las mismas condiciones, el resultado puede cambiar.**

Ejemplo:

*   Dos vuelos salen del mismo aeropuerto.
*   Misma aerolínea.
*   Misma hora.
*   Mismo día del año.

Uno puede retrasarse y el otro no.

Eso pasa porque:



*   El sistema aéreo tiene incertidumbre.
*   Hay eventos que no se pueden anticipar completamente.

De manera que no podemos decir:
*   Este vuelo se retrasará sí o sí.

Solo podemos decir:
*   Este vuelo **tiene mayor o menor probabilidad de retraso**


Machine Learning es ideal para estimar probabilidades, no certezas absolutas.

##**1.3. El efecto contextual de una variable.**

**Contextual** significa que una misma variable no tiene siempre el mimsmo efecto.

Ejemplo:

Volar a las 18:00

*   en un aeropuerto pequeño → puede no ser problema
*   en un hub congestionado → puede aumentar mucho el riesgo

Otro ejemplo:

El clima:

*   Una lluvia leve puede no afectar en un aeropuerto bien equipado.
*   la misma lluvia puede causar retrasos en otro aeropuerto.

El impacto de una variable depende del contexto completo del vuelo.

Esto es muy difícil de modelar con reglas fijas, pero ML lo aprende desde los datos.



*   Muchos factores influyen a la vez. Ninguna variable por sí sola explica el retraso operativo, si no la combinación de varias.

*   Existe incertidumbre inherente de qué factores se van a presentar.
*   Y si se presentan factores no se sabe con certeza su influencia en vuelo.

##**1.4. Resumen mental.**


**El problema:**

*   Tiene muchos factores interactuando
*   Tiene incertidumbre natural.
*   Depende del contexto específico de cada vuelo.

**Machine Learning:**

*   Aprende patrones complejos.
*   Combina variables automáticamente.
*   Estima riesgos probabilísticos.
*   Generaliza mejor que reglas manuales


**Conclusión:**

El retraso operativo de un vuelo es un fenómeno multifactorial, no determinista y contextual, ya que depende de la interacción de múltiples variables operacionales y temporales, presenta incertidumbre inherente y cuyo impacto varía según el contexto del vuelo. Por esta razón, resulta adecuado abordarlo mediante técnicas de Machine Learning.
