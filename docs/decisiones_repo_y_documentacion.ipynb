#**Decisiones sobre estructura y documentación del módulo Data Science.**

Este documento tiene como objetivo explicar y justificar las decisiones tomadas
respecto a la organización del repositorio y la forma de documentar el trabajo
del equipo de Data Science dentro del proyecto FlightOnTime.

El propósito es que todo el equipo comprenda estas decisiones y las utilice de
forma consistente durante el desarrollo del hackathon.

---

## **1. Estructura del repositorio de Data Science.**

Se definió una estructura de carpetas clara y simple, orientada a un MVP de
hackathon y alineada con buenas prácticas de Data Science en entornos reales.

La estructura busca:
- Separar datos originales de datos procesados
- Facilitar el trabajo colaborativo
- Evitar errores comunes (como modificar datos originales)
- Clarificar el rol del módulo de Data Science dentro del proyecto principal



### **Principales criterios**

- `data/raw/`  
  Contiene los datos originales sin ningún tipo de modificación.
  Esta carpeta se conserva como fuente única de verdad.

- `data/processed/`  
  Contiene copias de los datos ya limpios y preparados para el modelado.
  Todas las transformaciones se realizan sobre estos archivos.

- `notebooks/`  
  Espacio principal de trabajo del equipo de Data Science.
  Aquí se desarrolla el flujo completo: exploración, preparación y entrenamiento.

- `models/`  
  Contiene el modelo entrenado y serializado, listo para ser consumido por la API
  del equipo de Back-End.

- `src/`  
  Código reutilizable del proyecto (preprocesamiento, feature engineering y
  funciones auxiliares), con el objetivo de evitar duplicación en notebooks.

- `docs/`  
  Documentación de apoyo para explicar decisiones, definiciones y análisis.

- `images/`  
  Recursos visuales utilizados en documentación y presentaciones.

Esta estructura permite que cualquier integrante del equipo entienda rápidamente
dónde trabajar y qué rol cumple cada carpeta.


---

## **2. Decisiones sobre documentación.**

Se optó por una documentación **simple, centralizada y funcional**, evitando
sobrecargar el repositorio.

### **README del módulo Data Science.**

Se definió un único `README.md` en el submódulo de Data Science que incluye:
- Banner de identificación.
- Estado del proyecto.
- Descripción del submódulo.
- Explicación del uso de cada carpeta.

No se crearon README por carpeta para mantener el enfoque MVP y reducir ruido
documental.

### **Ubicación de los nombres del equipo.**

Los nombres de los integrantes del equipo se documentan únicamente en el
repositorio principal del proyecto.

Esto se decidió porque:
- El equipo pertenece al proyecto completo, no a un submódulo específico.
- Evita duplicación de información.
- Mantiene el README de Data Science enfocado en lo técnico.
