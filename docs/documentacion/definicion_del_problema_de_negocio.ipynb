#**1. Metodología para la definición del Problema de Negocio.**

El proyecto FlightOnTime tiene como objetivo **estimar de forma anticipada el riesgo de retraso operativo en vuelos comerciales**, utilizando datos históricos de la operación aérea.

Dado que la consigna del desafío no define explícitamente el concepto de retraso, el proyecto adopta una **definición operativa alineada con los estándares de la industria aeronáutica**, considerando el retraso como un evento medible y comparable bajo criterios estadístico-operacionales. En este contexto, el retraso se entiende como una desviación temporal relevante respecto del horario programado, susceptible de manifestarse en distintas etapas del proceso de vuelo y con impacto directo en la operación.

La estimación anticipada de este riesgo busca **apoyar la toma de decisiones de los distintos actores del ecosistema aeronáutico —tanto principales como afines—**, proporcionando una señal temprana que permita **gestionar, anticipar y mitigar los impactos operativos, económicos y reputacionales asociados a las demoras**, antes de que estas se materialicen en la ejecución real del vuelo.

#**2. Problema de Negocio.**

El proyecto FlightOnTime tiene como objetivo estimar de forma anticipada el riesgo de retraso operativo en vuelos comerciales, utilizando datos históricos de la operación aérea, con el fin de apoyar la toma de decisiones de los distintos actores del ecosistema aeronáutico —principales y afines— mediante una señal temprana que permita gestionar y mitigar el impacto de las demoras.
