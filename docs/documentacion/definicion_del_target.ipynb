#**Definición del Target – Retraso operativo en salida.**

##**1. Variable objetivo.**


*   `dep_op_delay` (binaria)

---



##**2. Regla de cálculo.**

Un vuelo se considera con **retraso operativo en salida** si el tiempo real de salida supera en 15 minutos o más la hora programada de salida.

*   **1 — Retrasado**: dep_delay ≥ 15
*   **0 — Puntual**: dep_delay < 15

---

##**3. Exclusiones explícitas.**


*   `cancelled` = 1
*   `diverted` = 1

Dado que:

*   un vuelo cancelado no operó, y
*   un vuelo desviado no llegó al destino programado,

ambos quedan fuera de la métrica de puntualidad operativa.


## **4. Interpretación del target.**



*   **Target = 1** representa la ocurrencia de un evento de retraso operativo en salida, medido bajo el estándar D15.
*   **Target = 0** representa una salida operativamente puntual.



El modelo estimará:

**La probabilidad de que un vuelo presente retraso operativo en su salida**, dadas las condiciones observables **antes del despegue**.
