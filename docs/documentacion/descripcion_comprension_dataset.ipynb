**6.2. Variables explicativas candidatas a features.**

Las variables explicativas corresponden a informaci√≥n disponible antes del despegue, por lo que pueden ser utilizadas para la predicci√≥n sin introducir fuga de informaci√≥n (data leakage).


Estas variables describen el contexto operacional, temporal y geogr√°fico del vuelo

**Variables temporales y de calendario.**

`year`,`month`,`day_of_month`,`day_of_week`,`fl_date`

Estas variables permiten capturar patrones asociados a estacionalidad, ciclos semanales y comportamiento temporal de la operaci√≥n a√©rea.

**Variables de programaci√≥n del vuelo.**

`crs_dep_time`, `crs_arr_time`,`crs_elapsed_time`, `distance`

Estas variables representan la planificaci√≥n original del vuelo y son conocidas antes de la operaci√≥n.

**Variables operativas y geogr√°ficas.**

`op_unique_carrier`,`op_carrier_fl_num`, `origin`,`origin_city_name`,`origin_state_nm`,`dest`,`dest_city_name`,`dest_state_nm`.


Estas variables permiten capturar diferencias estructurales entre aerol√≠neas, aeropuertos y rutas, que influyen en el riesgo de retraso.

**6.3. Variables excluidas del modelado.**

Las siguientes variables no deben utilizarse como features, ya que describen eventos posteriores al inicio de la operaci√≥n o explican directamente el retraso una vez ocurrido.

**Variables de estado del vuelo (solo para exclusi√≥n).**

`cancelled`, `diverted`, `cancellation_code`.


Estas variables se utilizan √∫nicamente para filtrar vuelos no operados, pero no forman parte del modelo predictivo.

**Variables temporales internas del proceso de vuelo.**

`dep_time`, `taxi_out`, `wheels_off`, `wheels_on`, `taxi_in`, `arr_time`, `actual_elapsed_time`, `air_time`.

**Variables causales del retraso.**


`carrier_delay`, `weather_delay`,`nas_delay`, `security_delay`, `late_aircraft_delay`.

Estas variables explican por qu√© ocurri√≥ el retraso y solo existen cuando el retraso ya se materializ√≥, por lo que su uso introducir√≠a label leakage.

Estas variables:
- Se generan **despu√©s de ocurrido el vuelo**.
- Explican **por qu√© ocurri√≥ el retraso**, no si ocurrir√°.
- No est√°n disponibles antes del despegue.

No deben utilizarse como variables explicativas ni para la definici√≥n del target, ya que su inclusi√≥n introducir√≠a data leakage, incorporando informaci√≥n que solo est√° disponible una vez materializado el retraso.
