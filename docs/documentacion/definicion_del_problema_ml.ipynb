#**Metodología para definir el problema de Machine Learning.**

A partir de a naturaleza del problema descrita en la sección anterior, se define el enfoque de Machine Learning considerando los siguientes criterios metodológicos:

##**1. Tipo de aprendizaje.**

El proyecto se aborda mediante **aprendizaje supervisado,** dado que se dispone de **registros históricos estructurados de vuelos** que incluyen tanto variables operativas como el resultado observado del vuelo (puntual o retrasado).

Estos datos etiquetados permiten al modelo aprender patrones a partir de ejemplos previos y generalizar su comportamiento a vuelos futuros.

##**2. Tipo de tarea.**

El problema se formula como una **tarea de clasificación,** ya que el objetivo no es predecir el tiempo exacto de retraso, sino **clasificar el vuelo según su estado operativo** bajo un estándar definido.

Esta formulación permite entregar una salida clara y accionable, alineada con el uso del modelo como herramienta de apoyo a la toma de decisiones.

##**3. Número de clases.**


Se adopta una **clasificación binaria,** coherente con la definición operativa del retraso:

**Clase 0 — Puntual:**
El vuelo presenta un retraso operativo inferior a 15 minutos.

**Clase 1 — Retrasado:**
El vuelo presenta un retraso operativo igual o superior a 15 minutos.

Esta definición transforma el fenómeno continuo del retraso en una señal binaria de riesgo, adecuada para un enfoque MVP.

##**4. Enfoque predictivo.**

El modelo se entrena para **estimar el riesgo de retraso antes de que el vuelo ocurra,** utilizando únicamente información disponible con anterioridad a la operación del vuelo. De este modo, la predicción conserva su utilidad práctica como herramienta anticipativa y evita el uso de información futura.

##**5. Relación con el negocio.**

La formulación del problema de Machine Learning mantiene una correspondencia directa con el problema de negocio, ya que la salida del modelo representa una **estimación probabilística del riesgo de retraso operativo,** y no una certeza absoluta, respetando la naturaleza incierta del fenómeno.
