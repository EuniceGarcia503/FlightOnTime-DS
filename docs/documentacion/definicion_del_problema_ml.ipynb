#**Definición del problema de Machine Learning.**

A partir de a naturaleza del problema de negocio descrita en la sección anterior, se define el enfoque de Machine Learning considerando los siguientes criterios metodológicos:

##**1. Tipo de tarea.**

El problema se formula como una **tarea de clasificación**, ya que el objetivo del modelo es asignar cada vuelo a una de dos categorías posibles: **vuelo puntual o vuelo retrasado**, según criterios operativos previamente definidos.

No se busca estimar un valor continuo de minutos de retraso, sino **clasificar el resultado operativo del vuelo**, lo que permite una interpretación simple y accionable para la toma de decisiones.

---

##**2. Tipo de aprendizaje.**

El enfoque adoptado corresponde a **aprendizaje supervisado**, dado que se dispone de **registros históricos de vuelos** que incluyen tanto las variables descriptivas del vuelo (aerolínea, aeropuerto, fecha, hora, distancia, entre otras) como el resultado observado de la operación, es decir, si el vuelo fue puntual o presentó retraso.

Estos datos históricos permiten entrenar un modelo que aprende la relación entre las condiciones del vuelo y su resultado operativo, para luego aplicar ese aprendizaje a vuelos futuros.

---

##**3. Definición.**

La combinación de una tarea de clasificación con aprendizaje supervisado resulta adecuada, ya que el problema cuenta con datos estructurados, resultados conocidos en el historial y un objetivo claramente definido desde la perspectiva del negocio.

De esta manera, el problema de Machine Learning queda definido de la siguiente forma:

**Construir un modelo de clasificación supervisada capaz de predecir si un vuelo comercial tendrá o no un retraso operativo, utilizando información histórica operacional y contextual para apoyar la toma de decisiones en la industria aérea.**
