#**Metodología para la definición del Problema de Negocio.**

#**1. Definición de retraso en el contexto de la industria aeronáutica.**

##**1.1. Registro de los retrasos.**

En la práctica operativa, **los retrasos de salida y de llegada se registran y analizan como eventos independientes.**

Un vuelo puede salir con retraso y recuperar tiempo durante el trayecto, o bien salir a tiempo y retrasarse durante la aproximación o el aterrizaje.

**Ejemplos:**

**Caso A:**
Sale con 20 minutos de retraso (salida impuntual), pero llega con solo 10 minutos de retraso (llegada puntual).

**Caso B:**
Sale puntualmente (0 minutos de retraso), pero debido al tráfico aéreo en el destino, aterriza 20 minutos tarde (llegada impuntual).

Estos casos muestran que el retraso puede manifestarse en distintas etapas del vuelo y afectar de forma distinta a la operación y a la experiencia del pasajero.

---

##**1.2. Uso del estándar de medición.**

El umbral de 15 minutos para definir un retraso corresponde a un estándar ampliamente adoptado en la industria aeronáutica para la medición de puntualidad operativa (On-Time Performance – OTP).

Bajo este estándar:

Salida (D15):
Un vuelo es puntual si el avión se retira de la puerta antes de que transcurran 15 minutos desde la hora programada.

Llegada (A15):
Un vuelo es puntual si el avión llega a la puerta de destino antes de que transcurran 15 minutos desde la hora programada.

Este criterio se utiliza principalmente para estadísticas, rankings de aerolíneas y análisis operativos, permitiendo comparar el desempeño entre diferentes transportistas y aeropuertos.


---

##**1.3. Diferencias entre umbral operativo y umbral legal.**

Es importante distinguir el estándar estadístico-operacional de ≥ 15 minutos de los umbrales legales utilizados para compensaciones a pasajeros, que suelen ser significativamente mayores y varían según el país.

**Ejemplo (contexto EE. UU.):**


| Escenario                     | Norma aplicable | Criterio         | Umbral                                  | Finalidad                           |
| ----------------------------- | --------------- | ---------------- | --------------------------------------- | ----------------------------------- |
| Medición estadística oficial  | 14 CFR Part 234 | Llegada          | ≥ 15 min                                | Medición de puntualidad (BTS / DOT) |
| Información al pasajero       | 14 CFR §259.8   | Salida o llegada | ≥ 30 min                                | Deber de información                |
| Derechos legales del pasajero | 14 CFR §260.2   | Salida o llegada | ≥ 3h (doméstico) / ≥ 6h (internacional) | Reembolsos y cambios contractuales  |

Cada definición es válida, pero responde a objetivos distintos, actores distintos y niveles de impacto distintos.

La clave del umbral ≥ 15 minutos es que:

*   No es un umbral legal.
*   Es un umbral estadístico-operacional.
*   Permite comparar desempeño operativo.


---

##**1.4. Vuelos desviados y vuelos cancelados.**


Además de los retrasos, la industria aeronáutica distingue explícitamente el tratamiento estadístico de los **vuelos desviados y vuelos cancelados,** ya que su inclusión directa en las métricas de puntualidad puede distorsionar el análisis operativo.



**Vuelos desviados.**

Un vuelo desviado es aquel que **aterriza en un aeropuerto distinto al originalmente programado**, debido a causas como condiciones meteorológicas adversas, emergencias médicas o problemas técnicos.

Desde el punto de vista estadístico:


*   **Tratamiento estándar:**
Los vuelos desviados no se consideran ni puntuales ni retrasados bajo la métrica A15.

*   **Exclusión de la métrica OTP:**
En la mayoría de los reportes de desempeño operativo, los vuelos desviados se excluyen del cálculo de puntualidad, ya que el avión nunca llegó a la puerta del destino original, condición necesaria para medir la llegada.


Este critério evita sesgos artificiales en los tiempos de llegada y preserva la comparabilidad entre aerolíneas y aeropuertos.

---

**Vuelos cancelados.**

Una cancelación no constituye un retraso. Conceptualmente:



* Recordemos que en un **retraso:** el vuelo se opera entre el origen y destino programado, pero fuera del horario previsto.
  
*   En la **Cancelación:** el vuelo no se realiza o se elimina de la programación, generalmente dentro de un horizonte previo a la salida.

Desde la perspectiva de las métricas operativas:

*   Impacto en la puntualidad (OTP):
Los vuelos cancelados no cuentan como “a tiempo”.
En el cálculo del porcentaje de puntualidad, se consideran como vuelos no realizados, lo que penaliza directamente el índice de desempeño de la aerolínea.

*   Métrica de regularidad:
La industria mide la Regularidad como el porcentaje de vuelos programados que efectivamente se operaron.
Es en esta métrica donde las cancelaciones tienen su mayor peso estadístico.

---

**Separación de métricas:**

El estándar global distingue claramente:






*   **Puntualidad:**
Vuelos a tiempo / Total de vuelos programados
(las cancelaciones reducen el denominador y afectan el resultado)
*   **Regularidad:**
Vuelos operados / Total de vuelos programados



*   **Retraso:**
Aplica únicamente a vuelos que despegaron y aterrizaron en los aeropuertos A y B programados, pero fuera del horario definido.

Desde la perspectiva de la experiencia del pasajero, una cancelación o un desvío puede percibirse como un “retraso masivo”; sin embargo, **la industria separa deliberadamente la tasa de retrasos de la tasa de cancelaciones** para permitir análisis operativos precisos y comparables.

---

##**1.5. Síntesis conceptual.**

El retraso puede manifestarse en distintas etapas del vuelo:

*   Antes del despegue (salida).
*   Durante el vuelo o la aproximación (llegada).



Ambos eventos retrasos en salida como retrasos en llegadas:

*   Se miden bajo el mismo umbral operacional, y
*   Son relevantes para distintos actores del ecosistema.

Desde la perspectiva de negocio, el retraso debe entenderse como un riesgo operativo asociado al vuelo como unidad, y no como un evento aislado.


Si el objetivo del proyecto es **apoyar decisiones operativas bajo estándares globales,** resulta necesario **separar conceptualmente los retrasos de los vuelos cancelados y desviados,** tratándolos como fenómenos distintos con métricas propias.

En consecuencia, el modelo de FlightOnTime se enfoca exclusivamente en el riesgo de retraso operativo, considerando únicamente vuelos efectivamente operados entre el origen y destino programados.

**Conclusión 1:**

Se adopta el concepto de “retraso operativo de vuelo” porque permite capturar, bajo un estándar reconocido por la industria, los impactos relevantes tanto para la operación aeroportuaria como para la experiencia del pasajero, manteniendo un enfoque general, comparable y escalable.

**Conclusión 2:**

Dado que el estándar de ≥ 15 minutos mide impacto operativo y de experiencia al pasajero, el alcance de la predicción beneficia:

*   Al núcleo del ecosistema aeronáutico (aeropuertos y aerolíneas).
*   A los pasajeros.
*   A actores de industrias relacionadas o dependientes del riesgo operativo del transporte aéreo.


Esta distinción es conceptual y no implica diseñar el modelo para un stakeholder específico, solo para posibles vistas en una capa interpretativa.



**Conclusión 3:**

La redacción final del problema de negocio se sustenta en los siguientes criterios:

*   Necesidad de negocio: estimar de manera anticipada el riesgo de demoras operativas utilizando registros históricos de vuelos.
*   Alcance de la predicción: apoyo a la toma de decisiones.
*   Actores beneficiados: aeropuertos, aerolíneas, pasajeros y actores relacionados.
*   Compatibilidad con baseline: no se especifica un contexto país, dado que el retraso operativo es un fenómeno universal.


---

#**2. Problema de Negocio.**

El proyecto FlightOnTime tiene como objetivo estimar de forma anticipada el riesgo de retraso operativo en vuelos comerciales, utilizando datos históricos de la operación aérea, con el fin de apoyar la toma de decisiones de los distintos actores del ecosistema aeronáutico —principales y afines— mediante una señal temprana que permita gestionar y mitigar el impacto de las demoras.
