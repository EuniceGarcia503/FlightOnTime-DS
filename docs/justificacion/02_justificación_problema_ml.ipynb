#**Justificación del problema de Machine Learning.**

Definir correctamente el problema de Machine Learning es un paso clave para asegurar que la **solución técnica esté alineada con la naturaleza real del fenómeno** que se desea modelar y con los objetivos del negocio.

---

##**Razonamiento 1:  Naturaleza multifactorial.**

El retraso de un vuelo no puede explicarse a partir de una única causa, sino que surge de la combinación de múltiples factores operativos y temporales, como el aeropuerto de origen, la aerolínea, la hora del día, el día de la semana, el mes del año, la estacionalidad y las condiciones meteorológicas. Ninguna de estas variables, por sí sola, resulta suficiente para explicar el comportamiento del vuelo; su efecto aparece cuando se **analizan de forma conjunta.**


---

##**Razonamiento 2: Carácter no determinista.**

Adicionalmente, el fenómeno presenta un carácter no determinista. **Vuelos que comparten condiciones muy similares pueden tener resultados distintos**: uno puede salir a tiempo y otro con retraso. Esto se debe a la incertidumbre inherente del sistema aeronáutico y a la existencia de eventos que no pueden anticiparse completamente. En consecuencia, no es posible definir reglas absolutas del tipo “este vuelo se retrasará”, sino únicamente estimar una mayor o menor probabilidad de retraso.

---

##**Razonamiento 3: El efecto contextual de una variable.**

Por último, el impacto de cada variable es contextual. **Una misma condición puede tener efectos distintos dependiendo del entorno operativo del vuelo**. Por ejemplo, una hora específica del día o una condición climática determinada puede no generar impacto en un aeropuerto, pero sí hacerlo en otro con mayor congestión o menor capacidad operativa. Este comportamiento es difícil de capturar mediante reglas fijas.

---

##**Razonamiento 4: Uso de Machine Learning.**



En conjunto, el retraso operativo de un vuelo es un fenómeno multifactorial, no determinista y contextual.


Por esta razón, el uso de **Machine Learning** resulta adecuado, ya que permite aprender **patrones complejos a partir de datos históricos de vuelos**, entendiendo cómo cada variable influye en el resultado operativo **en función del contexto en el que ocurre**. En este caso, el modelo aprende cómo factores como la aerolínea, el aeropuerto, la hora del día, el día de la semana o las condiciones operativas interactúan entre sí y **modifican el riesgo de que un vuelo sea clasificado como puntual o retrasado**.

De esta forma, el modelo no evalúa cada variable de manera aislada, sino que **captura el efecto conjunto y contextual de múltiples condiciones**, estimando una probabilidad de retraso operativo que refleja el comportamiento real del sistema aéreo. Esto permite representar la incertidumbre propia del proceso y alinear la predicción con el objetivo del negocio: anticipar riesgos operativos antes de que el vuelo se ejecute.
