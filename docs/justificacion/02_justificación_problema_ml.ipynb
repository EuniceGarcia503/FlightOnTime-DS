#**Justificación del problema de Machine Learning.**

##**1. El problema de negocio requiere Machine Learning.**

El problema de negocio identificado requiere un enfoque de Machine Learning debido a la naturaleza multifactorial, no determinista y contextual del retraso operativo de un vuelo.

##**1.1. Naturaleza multifactorial.**

El retraso de un vuelo **no puede explicarse a partir de una única causa, sino que surge de la combinación de múltiples factores operativos y temporales,** tales como el aeropuerto de origen, la aerolínea, la hora del día, el día de la semana, el mes del año, la estacionalidad, condiciones meteorológicas. Ninguna de estas variables, por sí sola, resulta suficiente para explicar el comportamiento del vuelo, sino que su impacto emerge de la interacción conjunta entre ellas.

##**1.2. Carácter no determinista.**

Adicionalmente, el fenómeno presenta un carácter no determinista, ya que vuelos que comparten **condiciones similares pueden tener resultados distintos.** Esto ocurre debido a la **incertidumbre inherente del sistema aeronáutico y a la presencia de eventos que no pueden anticiparse completamente.** En consecuencia, no es posible establecer reglas absolutas del tipo “este vuelo se retrasará”, sino únicamente estimar una mayor o menor probabilidad de retraso.

##**1.3. El efecto contextual de una variable.**

Finalmente, el impacto de cada variable es contextual: **una misma condición puede tener efectos distintos dependiendo del entorno operativo del vuelo.** Este comportamiento resulta difícil de capturar mediante reglas fijas, pero puede ser aprendido a partir de datos históricos mediante técnicas de Machine Learning.

**Conclusión:**

El retraso operativo de un vuelo es un fenómeno multifactorial, no determinista y contextual, ya que depende de la interacción de múltiples variables operacionales y temporales, presenta incertidumbre inherente y cuyo impacto varía según el contexto del vuelo. En este contexto, el uso de Machine Learning resulta adecuado para modelar patrones complejos, combinar múltiples variables y estimar riesgos probabilísticos, alineándose con la naturaleza del problema y con los objetivos del negocio.
