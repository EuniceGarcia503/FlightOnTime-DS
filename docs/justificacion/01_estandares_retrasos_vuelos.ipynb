#**Definiciones y estándares internacionales de reporte para vuelos retrasados, cancelados y desviados.**

**1. Estándar internacional de referencia – ICAO.**

La **Organización de Aviación Civil Internacional (ICAO)** establece los estándares y recomendaciones para la **recolección de datos de vuelos comerciales** a nivel global, incluyendo métricas y categorías técnicas que permiten comparar la operación de vuelos entre distintos países. Aunque ICAO define los criterios globales, estas definiciones no están publicadas abiertamente en su sitio web; sus manuales técnicos, como el *International Civil Aviation Vocabulary (Doc 9713)*, son de acceso restringido (Fuente: [ICAO – Aviation Data & Statistics](https://www.icao.int/aviation-data?utm_source=chatgpt.com)).

**2. Uso de referencias públicas basadas en estándares ICAO.**

Para efectos prácticos y de documentación verificable, se utilizan organizaciones nacionales y regionales que siguen los estándares ICAO, como:
- **FAA (Federal Aviation Administration) y BTS (Bureau of Transportation Statistics), EE. UU.**

- **EUROCONTROL** y regulaciones europeas como **EU Regulation 261/2004**.  
Estas entidades publican definiciones y estadísticas de vuelos retrasados, cancelados y desviados que siguen la lógica de estandarización internacional de ICAO (Fuente: [BTS – Major U.S. Air Carrier Delays, Cancellations and Diversions](https://www.bts.gov/content/major-us-air-carrier-delays-cancellations-and-diversions?utm_source=chatgpt.com)).

**3. Definiciones de vuelo adoptadas para análisis y documentación.**

- **Vuelo retrasado (Delayed flight):** llega a su puerta (gate) 15 minutos o más después de su hora programada de llegada (Fuente: [govinfo.gov – 14 CFR Part 234](https://www.govinfo.gov/content/pkg/CFR-2024-title14-vol4/html/CFR-2024-title14-vol4.htm)).

- **Vuelo cancelado (Cancelled flight):** vuelo programado que no se opera, pero estaba listado en el sistema de reservaciones dentro de los días previos a su salida (Fuente: [govinfo.gov – 14 CFR Part 234](https://www.govinfo.gov/content/pkg/CFR-2024-title14-vol4/html/CFR-2024-title14-vol4.htm)).

- **Vuelo desviado (Diverted flight):** vuelo que despega desde su origen previsto pero aterriza en un aeropuerto distinto al destino planificado (Fuente: [govinfo.gov – 14 CFR Part 234](https://www.govinfo.gov/content/pkg/CFR-2024-title14-vol4/html/CFR-2024-title14-vol4.htm)).

**4. Estándares de reporte y códigos de eventos (BTS Form 234).**

Las aerolíneas reportan mensualmente a la **BTS / FAA** los datos de operación según el **BTS Form 234 “On-Time Flight Performance Report”**, siguiendo los estándares de 14 CFR Part 234 (Fuente: [bts.gov](https://www.bts.gov/topics/airlines-and-airports/number-15-gate-departure-time?utm_source=chatgpt.com)).


**Códigos de cancelación.**

Estos códigos describen por qué una aerolínea canceló un vuelo:
- **A – Air Carrier:** Cancelación por razones asociadas a la aerolínea (equipos, tripulación, etc.).

- **B – Weather / Extreme Weather:** Cancelación por condiciones climáticas extremas.

- **C – National Aviation System (NAS):** Cancelación debido a factores del sistema nacional de aviación (congestión, control de tráfico, etc.).

- **D – Security:** Cancelación por razones de seguridad.  
Estos códigos aparecen en los datos oficiales bajo el campo **Cancellation Code** para cada vuelo reportado (Fuente: [bts.gov](https://www.bts.gov/topics/airlines-and-airports/number-15-gate-departure-time?utm_source=chatgpt.com)).

**Códigos de retraso.**

Para vuelos que llegaron tarde (≥15 min), se reportan causas según otro conjunto de códigos:
- **E – Carrier Caused:** retraso por aerolínea.

- **F – Weather:** retraso por condiciones climáticas.

- **G – NAS:** retraso por Sistema Nacional de Aviación.

- **H – Security:** retraso por razones de seguridad.

- **I – Late Arriving Aircraft:** retraso por avión que llegó tarde desde vuelo anterior.  

Estos códigos se usan en el reporte para indicar la causa principal del retraso cuando aplica (Fuente: [bts.gov](https://www.bts.gov/topics/airlines-and-airports/number-15-gate-departure-time?utm_source=chatgpt.com)).

**5. Registro de vuelos cancelados y desviados.**

- **Cancelados:** se registran con horas reales de salida y llegada en cero y con un código de cancelación (A–D).  

- **Desviados:** se registran con tiempo de salida real pero con llegada en cero, manteniendo el número de vuelo original.

**6. Justificación del uso de estos estándares.**

- Todas las definiciones y códigos reportados están documentados y verificados por fuentes oficiales, como BTS / FAA, que actúan bajo estándares ICAO.

- Permiten **documentar y analizar el desempeño operativo de los vuelos comerciales** de manera consistente.  

- Son ampliamente aceptadas en la industria y en estudios académicos internacionales.
