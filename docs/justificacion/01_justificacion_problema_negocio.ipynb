## **Justificación del Problema de Negocio**

La definición de problema de negocio constituye un paso previo indispensable para la formulación del problema de Machine Learning, al establecer con claridad la necesidad de negocio que se busca satisfacer y la importancia de abordarla mediante un modelo predictivo.

## **Decisión 1: Interpretación del desafío planteado.**

El proyecto FlightOnTime se desarrolla en el marco del Hackathon ONE II – Latam, cuyo desafío general **plantea el desarrollo de un modelo capaz de estimar si un vuelo va a despegar a tiempo o con retraso.**

Más allá de su formulación explícita, el desafío refleja una necesidad implícita del negocio:
**anticipar el riesgo de demoras en vuelos comerciales** y utilizar dicha estimación como una **herramienta de apoyo a la toma de decisiones**, orientada a **mitigar el impacto operativo, económico y reputacional de los retrasos** sobre los distintos actores del ecosistema aeronáutico.

## **Decisión 2: Identificar la ambigüedad del concepto “retraso”**

La consigna del desafío **no define de manera explícita qué se entiende por “retraso”**, ni bajo qué **estándar debe medirse dicho evento.**

Esta ambigüedad inicial representa un **punto crítico para el diseño de la solución,** ya que distintas interpretaciones del concepto de retraso podrían conducir a resultados inconsistentes y poco alineados con las prácticas reales de la industria aeronáutica.

En este contexto, el equipo identifica la necesidad de **definir explícitamente qué se considera un retraso** antes de traducir el problema de negocio a un problema de Machine Learning, estableciendo un **criterio claro y documentado** que permita:

Evitar interpretaciones arbitrarias.

Alinear la solución con estándares utilizados en la industria aeronáutica.

Construir una **baseline reutilizable y escalable** para futuras extensiones del proyecto.
