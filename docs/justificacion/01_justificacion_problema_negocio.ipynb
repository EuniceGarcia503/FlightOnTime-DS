## **Justificación del Problema de Negocio**

La definición de problema de negocio constituye un paso previo indispensable para la formulación del problema de Machine Learning, al establecer con claridad la necesidad de negocio que se busca satisfacer y la importancia de abordarla mediante un modelo predictivo.

---

## **Decisión 1: Interpretación del desafío planteado.**

El proyecto FlightOnTime se desarrolla en el marco del Hackathon ONE II – Latam, cuyo desafío general **plantea el desarrollo de un modelo capaz de estimar si un vuelo va a despegar a tiempo o con retraso.**

Más allá de su formulación explícita, el desafío refleja una necesidad implícita del negocio:
**anticipar el riesgo de demoras en vuelos comerciales** y utilizar dicha estimación como una **herramienta de apoyo a la toma de decisiones**, orientada a **mitigar el impacto operativo, económico y reputacional de los retrasos** sobre los distintos actores del ecosistema aeronáutico.


---

## **Decisión 2: Identificar la ambigüedad del concepto “retraso”**

La consigna del desafío **no define de manera explícita qué se entiende por “retraso”**, ni bajo qué **estándar debe medirse dicho evento.**

Esta ambigüedad inicial representa un **punto crítico para el diseño de la solución,** ya que distintas interpretaciones del concepto de retraso podrían conducir a resultados inconsistentes y poco alineados con las prácticas reales de la industria aeronáutica.

En este contexto, el equipo identifica la necesidad de **definir explícitamente qué se considera un retraso** antes de traducir el problema de negocio a un problema de Machine Learning, estableciendo un **criterio claro y documentado** que permita:

Evitar interpretaciones arbitrarias.

Alinear la solución con estándares utilizados en la industria aeronáutica.

Construir una **baseline reutilizable y escalable** para futuras extensiones del proyecto.


---

## **Decisión 3: Definición operativa del retraso.**

###**Registros de los retrasos**

En la práctica operativa de la industria aeronáutica, **los retrasos de salida y de llegada se registran y analizan como eventos independientes.**

Un vuelo puede salir con retraso y recuperar tiempo durante el trayecto, o bien salir a tiempo y retrasarse durante la aproximación o el aterrizaje. Por esta razón, la puntualidad en salida y la puntualidad en llegada responden a dinámicas operativas distintas y no son directamente intercambiables.

**Ejemplos:**




*   **Caso A:** El vuelo sale con 20 minutos de retraso (salida impuntual), pero logra recuperar tiempo en ruta y llega con solo 10 minutos de retraso (llegada puntual).

*   **Caso B:** El vuelo sale puntualmente (0 minutos de retraso), pero debido a congestión aérea en el aeropuerto de destino, aterriza con 20 minutos de retraso (llegada impuntual).

Estos casos ilustran que el retraso puede manifestarse en distintas etapas del vuelo y afectar de manera diferente tanto a la operación como a la experiencia del pasajero.

---

###**Vuelos desviados y cancelados**

Además de los retrasos, la industria aeronáutica distingue explícitamente el tratamiento estadístico de **vuelos desviados y vuelos cancelados**, ya que su inclusión directa en las métricas de puntualidad puede distorsionar el análisis operativo.

**Vuelos desviados**

Un vuelo desviado es aquel que aterriza en un aeropuerto distinto al originalmente programado, debido a causas como condiciones meteorológicas adversas, emergencias médicas o problemas técnicos.

Desde el punto de vista estadístico:

*   **Tratamiento estándar:** Los vuelos desviados no se consideran ni puntuales ni retrasados bajo la métrica de llegada, ya que no cumplen la condición de arribar al aeropuerto de destino original.
*   **Exclusión de la métrica OTP:** En la mayoría de los reportes de desempeño operativo, los vuelos desviados se excluyen del cálculo de puntualidad, dado que no existe un tiempo de llegada válido a la puerta de destino programada.

Este criterio evita sesgos artificiales en los indicadores de puntualidad y preserva la comparabilidad entre aerolíneas y aeropuertos.

---

###**Vuelos cancelados**

Una cancelación no constituye un retraso. Conceptualmente:


*   En un **retraso**, el vuelo se opera entre el origen y el destino  programado, pero fuera del horario previsto.
*   En una **cancelación**, el vuelo no se realiza o se elimina de la programación, generalmente con anterioridad a la salida.

Desde la perspectiva de las métricas operativas:



*   **Impacto en la puntualidad (OTP):**
Los vuelos cancelados no se contabilizan como “a tiempo”. En los indicadores de puntualidad, se consideran como vuelos no realizados, lo que penaliza el desempeño global de la aerolínea.
*   **Métrica de regularidad:**
La industria mide la regularidad como el porcentaje de vuelos programados que efectivamente se operaron. Es en esta métrica donde las cancelaciones tienen su mayor peso estadístico.

## **Decisión 4: Adoptar un estándar de la industria”**

El umbral de 15 minutos para definir un retraso corresponde a un estándar ampliamente adoptado en la industria aeronáutica para la medición de la puntualidad operativa (On-Time Performance – OTP).

Bajo este estándar:

Salida (D15):
Un vuelo se considera puntual en su salida si el avión se retira de la puerta antes de que transcurran 15 minutos desde la hora programada.

Llegada (A15):
Un vuelo se considera puntual en su llegada si el avión llega a la puerta de destino antes de que transcurran 15 minutos desde la hora programada.

Este criterio se utiliza principalmente para estadísticas oficiales, rankings de aerolíneas y análisis operativos, permitiendo comparar el desempeño entre distintos aerolíneas y aeropuertos de manera consistente.

---

**Diferencia entre umbral operativo y umbrales legales**

Es importante distinguir el estándar estadístico–operacional de ≥ 15 minutos de los umbrales legales utilizados para compensaciones a pasajeros, los cuales suelen ser significativamente mayores y varían según el país.


| Escenario                     | Norma aplicable | Criterio         | Umbral                                  | Finalidad                           |
| ----------------------------- | --------------- | ---------------- | --------------------------------------- | ----------------------------------- |
| Medición estadística oficial  | 14 CFR Part 234 | Llegada          | ≥ 15 min                                | Medición de puntualidad (BTS / DOT) |
| Información al pasajero       | 14 CFR §259.8   | Salida o llegada | ≥ 30 min                                | Deber de información                |
| Derechos legales del pasajero | 14 CFR §260.2   | Salida o llegada | ≥ 3h (doméstico) / ≥ 6h (internacional) | Reembolsos y cambios contractuales  |


Cada definición es válida, pero responde a objetivos distintos, actores distintos y niveles de impacto distintos.

---

**Clave del umbral ≥ 15 minutos**

El uso del umbral ≥ 15 minutos se justifica porque:

*   Está alineado con prácticas reales de la industria.
*   No es un umbral legal.
*   Es un umbral estadístico-operacional.
*   Permite comparar desempeño operativo.
